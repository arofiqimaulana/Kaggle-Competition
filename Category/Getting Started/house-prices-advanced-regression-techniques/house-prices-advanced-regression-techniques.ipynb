# Handle Missing Values
Cara Handle missing values dapat melalui
1. Average Value Method
2. Regression Method
3. ignore observation

Kita akan bagi cara handle missing values berdasarkan
1. Numerical Variable
2. Categorical Variabel

## 1. Handle Missing Values for Numerical Variable

<img src="images/house-prices-advanced-regression-techniques.png"/>

```python
def HandleMissingValuesNumericalVar(variabel_name):
    from sklearn.linear_model import LinearRegression
    
    # Bagi data menjadi dua bagian besar yaitu null value dan not null value
    full_data = df_train[[variabel_name,"SalePrice"]]
    null_data = full_data[full_data[variabel_name].isnull()]
    notnull_data = full_data[~(full_data[variabel_name].isnull())]
    
    # Estimasi missing values via regresi
    x = np.array(notnull_data[variabel_name]).reshape(-1,1)
    y = np.array(notnull_data["SalePrice"])
    model = LinearRegression()
    fit = model.fit(x,y)
    b = fit.coef_
    a = fit.intercept_
    predict_x = (np.array(null_data["SalePrice"])-a)/b
    indeks = null_data.index.values.tolist()
    df_predict_x = pd.DataFrame(predict_x,index=indeks,columns=[variabel_name])
    df_y = pd.DataFrame(null_data["SalePrice"])
    
    #Create New Dataframe
    new_null_df = pd.concat([df_predict_x,df_y],ignore_index=True,axis=1)
    new_null_df.columns=[variabel_name,"SalePrice"]
    
    # Merge
    df_new = pd.concat([notnull_data,new_null_df]).sort_index()
    var = pd.DataFrame(df_new[variabel_name],columns=[variabel_name])

    return var
```

```python
df_LotFrontage = HandleMissingValuesNumericalVar("LotFrontage")
df_LotFrontage.head()
```

## 2. Handle Missing Values for Categorical Variable

```python
def HandleMissingValuesCategoricalVar(variable_name):
    from sklearn.linear_model import LogisticRegression
    
    # Bagi data menjadi dua bagian yaitu null_data dan non_null_data
    full_data=df_train[[variable_name,"SalePrice"]]
    null_data = full_data[full_data[variable_name].isnull() | (full_data[variable_name]=='None')]
    not_null_data = full_data[~(full_data[variable_name].isnull() | (full_data[variable_name]=='None'))]
    
    # Estimasi via regresi logistik
    y = np.array(not_null_data[variable_name])
    x = np.array(not_null_data['SalePrice']).reshape(-1,1)
    
    model = LogisticRegression()
    fit = model.fit(x,y)
    predict = fit.predict(np.array(null_data["SalePrice"]).reshape(-1,1))
    
    # Data Cleaning
    indeks = null_data.index.values.tolist()
    df_predict = pd.DataFrame(predict,index=indeks)
    y_null = pd.DataFrame(null_data["SalePrice"])  
    
    #Create New Dataframe
    new_null_df = pd.concat([df_predict,y_null],ignore_index=True,axis=1)
    new_null_df.columns = [variable_name,"SalePrice"]
    
    # Merge Data
    df_new = pd.concat([not_null_data,new_null_df]).sort_index()
    var = pd.DataFrame(df_new[variable_name],columns=[variable_name])

    return var
```

```python
df_Alley = HandleMissingValuesCategoricalVar("Alley")
df_Alley.head()
```

# 3. Handle Missing Values to all variable

```python
null_var = df_train.isna().sum()[df_train.isna().sum() != 0].index.values.tolist()
cat_var = df_train.dtypes[df_train.dtypes == 'object'].index.values.tolist()
num_var = df_train.dtypes[df_train.dtypes != 'object'].index.values.tolist()

cat_null_var = [k for k in cat_var if k in null_var ]
num_null_var = [k for k in num_var if k in null_var ]
```

```python
df = []
for k in num_null_var:
    df.append(HandleMissingValuesNumericalVar(k))
df_new_null_num_var = pd.concat([df[0],df[1],df[2]],axis=1)
df_new_null_num_var
```

```python
df = []
for k in cat_null_var:
    df.append(HandleMissingValuesCategoricalVar(k))
df_new_null_cat_var = pd.concat([df[0],df[1],df[2],df[3],df[4],
                                df[5],df[6],df[7],df[8],
                                df[9],df[10],df[11],df[12],
                                df[13],df[14],df[15]],axis=1)
df_new_null_cat_var
```

# 5. Merge Data

```python
null_var = cat_null_var + num_null_var
```

```python
var_lain = [k for k in df_train.columns.values.tolist() if k not in null_var ]
```

```python
ALLDF = pd.concat([df_new_null_num_var,df_new_null_cat_var,df_train[var_lain]],axis=1)
ALLDF
```
